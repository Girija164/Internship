# 2.Python program to scrap The product details from amazon.in

# To Find Tag and scrap the text out of it

```python
brand_name=[]
product_name=[]
return_exchange=[]
expected_delivery = []
availability=[]
product_url=[]
price=[]
```

# To Get URLS To Open Them

```python
urls=[]
# To scrape 3 pages
for i in range(0,3):
   
    url_tag= driver.find_elements_by_xpath("//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
    price_tag= driver.find_elements_by_xpath("//a[@class='a-size-base a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
    for url in url_tag:
        urls.append(url.get_attribute('href'))
    for prices in price_tag:
        price.append(prices.text.split())
    
    #To find next is clickable
    while True:
        try:
            nxt=driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").click()
            time.sleep(2)
        except NoSuchElementException:
            break
    
    
```

```python
# To open each url
urls1= urls[0:100]
for i in urls1:
    driver.get(i)
    time.sleep(2)

#Brand Name

    try:
        brand_tag=driver.find_element_by_xpath("//td[@class='a-span9']/span")
        brand_name.append(brand_tag.text)
    except NoSuchElementException:
        brand_name.append('-')
        
#Product Name

    try:
        product_tag= driver.find_element_by_id('productTitle')
        product_name.append(product_tag.text)
    except NoSuchElementException:
        product_name.append('-')
        
#Return/Exchange

    try:
        exchange=driver.find_element_by_xpath("//a[@class='a-size-small a-link-normal a-text-normal']")
        return_exchange.append(exchange.text)
    except NoSuchElementException:
        return_exchange.append('-')

#Expected Delivery

    try:
        delivery=driver.find_element_by_xpath("//div[@class='a-spacing-base']/b")
        expected_delivery.append(delivery.text)
    except NoSuchElementException:
        expected_delivery.append('-')
        
#Availability

    try:
        avail=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none }']/span")
        availability.append(avail.text)
    except NoSuchElementException:
        availability.append('-')
    
```

```python
price=price[0:69]
product_url=urls1
```

```python
print(len(brand_name), len(product_name), len(return_exchange), len(expected_delivery), len(availability),len(product_url), len(price))
```

# Make DataFrame

```python
df=pd.DataFrame({'Brand':brand_name,
                 'Product_Name' :product_name,
                 'Return/Exchange':return_exchange,
                 'Delivery': expected_delivery,
                 'Availbility': availability,
                 'URL':product_url,
                 'Price':price})
df
```

```python
df.to_csv('amazon.csv')
```

```python
#To find facilities
facilities=[]
try:
    fac_tag = driver.find_elements_by_xpath("//div[@class='has-wifi']")
    for i in fac_tag:
        facilities.append(i.text)
except NoSuchElementException:
    facilities.append('-')
```

```python
facilities1=[]
try:
    fac1_tag = driver.find_elements_by_xpath("//div[@class='has-sanitation']")
    for i in fac1_tag:
        facilities1.append(i.text)
except NoSuchElementException:
    facilities1.append('-')
```

```python
url=[]
try:
    url_tag = driver.find_elements_by_xpath("//h2[@class='title title-6']//a")
    for i in url_tag:
        url.append(i.get_attribute('href'))
except NoSuchElementException:
    url.append('-')
```

```python
urls=url
```

```python
prop_des=[]
for i in urls:
    driver.get(i)
    time.sleep(2)
    
    #To find property Description
    try:
        desc= driver.find_elements_by_xpath("//div[@class='content']")
        for dsc in desc:
            prop_des.append(dsc.text)
    except NoSuchElementException:
        prop_des.append('-')  
    
      
```

```python
prop_des=prop_des[6::16]
```
