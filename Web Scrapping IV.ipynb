# 5. Scrape the details of trending repositories on Github.com. Url = https://github.com/
You have to find the following details:
    
A) Repository title

B) Repository description

C) Contributors count

D) Language used

Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver= webdriver.Chrome(r'C:\Users\user\AppData\Local\Temp\Rar$EXa0.836\chromedriver.exe')
```

```python
url='https://github.com/trending'
```

```python
driver.get(url)
```

```python
driver.maximize_window()
time.sleep(2)
```

```python
url = driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a")

urls = []
for i in url:
    urls.append(i.get_attribute('href'))
```

```python
repo_title=[]
description=[]
contributors=[]
lang=[]

for i in urls:
    driver.get(i)
    #repository
    try:
        repository_tag = driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a")
        repo_title.append(repository_tag.text)
    except NoSuchElementException:
        repo_title.append("-")
        
    #description
    try:
        description_tag= driver.find_element_by_xpath("//div[@class='BorderGrid-cell']/p")
        description.append(description_tag.text)
    except NoSuchElementException:
        description.append("-")
        
    #contributors
    try:
        contributors_tag = driver.find_element_by_xpath("//h2[@class='h4 mb-3']/a[contains(text(),'Contributors')]/span")
        contributors.append(contributors_tag.text)
    except NoSuchElementException:
        contributors.append("-")
  
    
    #languages 
    try:
        languages=driver.find_elements_by_xpath("//li[@class='d-inline']/a/span[1]")
        for i in languages:
            lang.append(i.text.split('\n')[0])  
    except NoSuchElementException:
        lang.append("-")
```

```python
lang=lang[0:25]
```

```python
print(len(repo_title), len(description), len(contributors), len(lang))
```

```python
df=pd.DataFrame({
        'Repository_title': repo_title,
        'Description': description,
        'Contributors_count': contributors,
        'Language_used': lang
    })
df
```

```python
driver.close()
```

```python
#To get type
types=[]
try:
    type_tag= driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[2]/p")
    for i in type_tag[1:]:
        types.append(i.text)
except NoSuchElementException:
    types.append("-")
```

```python
#To get task
task=[]
try:
    task_tag= driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[3]/p")
    for i in task_tag[1:]:
        task.append(i.text)
except NoSuchElementException:
    task.append("-")
```

```python
#To get Attribute_type
att_type=[]
try:
    att_tag= driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[4]/p")
    for i in att_tag[1:]:
        att_type.append(i.text)
except NoSuchElementException:
    att_type.append("-")
```

```python
#To get instance
instance=[]
try:
    instance_tag= driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[5]/p")
    for i in instance_tag[1:]:
        instance.append(i.text)
except NoSuchElementException:
    instance.append("-")
```

```python
#To Get Attribute
attribute=[]
try:
    attribute_tag= driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[6]/p")
    for i in attribute_tag[1:]:
        attribute.append(i.text)
except NoSuchElementException:
    attribute.append("-")
```

```python
#To get Year
year=[]
try:
    year_tag= driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[7]/p")
    for i in year_tag[1:]:
        year.append(i.text)
except NoSuchElementException:
    year.append("-")
```

```python
print(len(name), len(types), len(task), len(att_type), len(instance), len(attribute), len(year))
```
